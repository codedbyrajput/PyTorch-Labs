```python
# Distances in miles for recent bike deliveries
#dtype=torch.float32 sets your data type to 32-bit floating point values for
#precise calculations.
distances = torch.tensor([[1.0], [2.0], [3.0], [4.0]], dtype=torch.float32)

# Corresponding delivery times in minutes
times = torch.tensor([[6.96], [12.11], [16.77], [22.21]], dtype=torch.float32)
```

Every neuron is trained using linear regression (y = mx + b).

These neurons are trained simulataneously as a team. The data goes to first layer of neurons and it does it math and then passes to the next layer and then the final layer gives out the prediction.

If the prediction is wrong then it sees how much each neuron is to blame for the wrong output hence it changes the weight and biases at the exact same time this process is called Backpropagation.



---

Now, lets say that one layer of neuron did its work and pass it to the next layer, then it might end up passing the linear regression
which will end up giving a vrey complex equation of a single line. To prevent this from happening we have activation functions inside every neuron. This activation function adds a bit of bend to our equation before passing it to the next layer to not make our data boring and helps to model to learn complex data.

---

Three big activation functions

1. ReLU (Rectified Linear Unit): The "Bouncer."

The Rule: If the raw score is negative, the output is exactly 0. If the raw score is positive, it passes the exact number through unchanged.

Why it's used: It is incredibly fast for computers to calculate, making it the default choice for the hidden "middle" layers of most modern neural networks.

2. Sigmoid: The "Probability Squisher."

The Rule: It takes any number, no matter how huge or tiny, and squishes it into a gentle curve between 0 and 1.

Why it's used: Because the output is always between 0 and 1, it is perfect for the final output layer when you need the model to give you a percentage or probability (e.g., "There is an 85% chance this delivery will be on time").

3. Tanh (Hyperbolic Tangent): The "Zero-Centered Squisher."

The Rule: Very similar to Sigmoid, but it squishes numbers between -1 and 1.

Why it's used: It helps keep the data centered around zero, which can sometimes help the network learn faster during training than Sigmoid.



```python
# Create a model with one input (distance) and one output (time)
"""
nn.Linear(1,1) creates a linear model. Here, the first 1 means it takes one
input (distance) and the second one means it gives one output (predicted time).
This single linear Layer will automatically manage the weights and biases for us.
If I have 2,1 paramters then it means my model should expect 2 inputs and give
1 output for eg: 2 variables can be distance and weather.
"""
model = nn.Sequential(nn.Linear(1, 1))
```

Now to train the model, we need to use 2 key tools to help modoel to learn from the data:

1. Loss function: nn.MSELoss defines Mean Squared Error loss function. It will measure how wrong our predictions are and will try to minimize this error.
2. Optimizer: Loss function can only tell how bad the model messed up thus we need optimizer to fix that. They help to readjust the weights and bias to reduce MSE. W(new) = W(old) - (lr x Gradient). Here this formula helps to readjust. Gradient is the slope like in multivariable calculus, lr is the learning rate, it dictates how big of a step we need to take to reach a certain point in training. If lr is too small then we would need massive computation time to reach the end of the valley. If its too big then the model might miss the bottom completley and land on the other side of the and bounce out of control this is called overshooting.

```python
# Define the loss function and optimizer
loss_function = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

Upon doing nn.linear(1,1) we generated a random number for both W and B in the linear equation but now to train our model and get the best fitted regression line we need to change these random values.

1. optimizer.zero_grad(): erasing the whiteboard. delete the previosu values.

2. outputs = model(distances): take the practice test. This is forward pass. we look at the distance and make out guess with currently what we know about the values of W and B.

3. loss = loss_function(outputs, times): Grade the test. We check the predicted output with the actual output.

4. loss.backward(): review mistake. This is called backpropagation. It goes back and adjusts the weight and bias accordingly to improve the model.

5. optimizer.step(): Studying/Updating the Brain. NOw we know what was did wrong. The optimizer reaches inot the model and adjusts the weights and biases slighlty so that the next time it sees those same distances, it will guess the time a bit closer to reality.

```python
# Training loop
for epoch in range(500):
    # Reset the optimizer's gradients
    optimizer.zero_grad()
    # Make predictions (forward pass)
    outputs = model(distances)
    # Calculate the loss
    loss = loss_function(outputs, times)
    # Calculate adjustments (backward pass)
    loss.backward()
    # Update the model's parameters
    optimizer.step()
    # Print loss every 50 epochs
    if (epoch + 1) % 50 == 0:
        print(f"Epoch {epoch + 1}: Loss = {loss.item()}")
```
